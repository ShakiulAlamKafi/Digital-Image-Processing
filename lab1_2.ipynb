```python
import PIL.Image as image
import matplotlib.pyplot as plt
```

```python
# Read an image.
rgb_img = image.open("rgmsc_logo.jpeg")

plt.imshow(rgb_img)
plt.show()
```

```python
_mode = rgb_img.mode
print("Color mode is: ", _mode)
```

```python
print(rgb_img.size)
```

```python
width, height = rgb_img.size
print(f"Resolution = {width} x {height} pixels")
```

```python
gray_img = rgb_img.convert('L')
plt.imshow(gray_img, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(15,8)) #inch
plt.subplot(1,2,1)
plt.imshow(rgb_img)
plt.title("original image")

plt.subplot(1,2,2)
plt.imshow(gray_img, cmap='gray')
plt.title("grayscale image")

plt.show()
```

```python
gray_img.save("grayscale.jpg")
```

```python
import numpy as np

img = np.array(gray_img).astype('float32')
s1 = 1.5
s2 = 0.3

bright_img = img * s1
bright_img = np.clip(bright_img, 0, 255).astype('uint8')

dark_img = img * s2
dark_img = np.clip(dark_img, 0, 255).astype('uint8')


plt.figure(figsize=(15,8)) #inch
plt.subplot(1,3,1)
plt.imshow(gray_img, cmap='gray')
plt.title("grayscale image")

plt.subplot(1,3,2)
plt.imshow(bright_img, cmap='gray')
plt.title("brighter grayscale image")

plt.subplot(1,3,3)
plt.imshow(dark_img, cmap='gray')
plt.title("darker grayscale image")

plt.show()
```
