Lab-2: Brightness adjustment of grayscale images

library

```python
import PIL.Image as image
import matplotlib.pyplot as plt
import numpy as np
```

1.	Read a grayscale image.

```python
gray_img=image.open('rglhs_logo.png')
plt.imshow(gray_img, cmap='gray')
plt.show()
```

2.	Increase the brightness of the image by multiplying each pixel value by a scaling factor S₁ = 1.5

```python
img=np.array(gray_img)
img=img.astype(np.float32)
s1=1.5
bright_img=s1*img
bright_img=np.clip(bright_img,0,255)
bright_img=bright_img.astype(np.uint8)
plt.imshow(bright_img, cmap='gray')
plt.show()
```

3.	Decrease the brightness of the image by multiplying each pixel value by a scaling factor S₂ = 0.3

```python
s2=0.3
dark_img=s2*img
dark_img=np.clip(dark_img,0,255)
dark_img=dark_img.astype(np.uint8)
plt.imshow(dark_img, cmap='gray')
plt.show()
```

4.	Compare the original grayscale image with the brightness-adjusted images.

```python
plt.figure(figsize=(15, 8))
plt.subplot(1, 3, 1)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')
plt.subplot(1, 3, 2)
plt.imshow(bright_img, cmap='gray')
plt.title('Brightness increased Grayscale Image')
plt.subplot(1, 3, 3)
plt.imshow(dark_img, cmap='gray')
plt.title('Brightness decreased Grayscale Image')
plt.show()
```
