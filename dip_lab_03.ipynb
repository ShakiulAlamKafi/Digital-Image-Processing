Lab-3: Image Thresholding

```python
import PIL.Image as image
import numpy as np
import matplotlib.pyplot as plt
```

a) Read a grayscale image

```python
img = image.open("VU-Logo-gray.png")
```

```python
plt.imshow(img, cmap="gray")
plt.show()
```

b) Set a threshold value, k = 128

```python
k = 128
```

c) Apply binary thresholding using:
     if r > k → s = 255
     else → s = 0


```python
img_array = np.array(img)
```

```python
img_array.shape
```

```python
binary_128 = np.zeros_like(img_array)

for i in range(img_array.shape[0]):
    for j in range(img_array.shape[1]):
        if img_array[i, j] > k:
            binary_128[i, j] = 255
        else:
            binary_128[i, j] = 0
```

d) Display the binary image.

```python
plt.imshow(binary_128, cmap="gray")
plt.show()
```

e) Create three binary images using different thresholds: k = 50, 100, 150.

```python
def fun(arr, k):
    binary_img = np.zeros_like(arr)

    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]):
            if arr[i, j] > k:
                binary_img[i, j] = 255
            else:
                binary_img[i, j] = 0
    return binary_img
```

```python
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
binary_128=fun(img_array, 128)
plt.imshow(binary_128, cmap="gray")
plt.title("Threshold = 128")
plt.axis("off")

plt.subplot(2,2,2)
binary_50=fun(img_array, 50)
plt.imshow(binary_50, cmap="gray")
plt.title("Threshold = 50")
plt.axis("off")

plt.subplot(2,2,3)
binary_100=fun(img_array, 100)
plt.imshow(binary_100, cmap="gray")
plt.title("Threshold = 100")
plt.axis("off")

plt.subplot(2,2,4)
binary_150=fun(img_array, 150)
plt.imshow(binary_150, cmap="gray")
plt.title("Threshold = 150")
plt.axis("off")

plt.tight_layout()
plt.show()
```
