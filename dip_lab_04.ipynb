Lab-4: Image Negative Transformation

```python
import PIL.Image as image
import numpy as np
import matplotlib.pyplot as plt
```

a) Read a grayscale image

```python
img = image.open("hand_xray.jpg")
plt.imshow(img, cmap="gray")
plt.show()
```

b) Print the intensity values of a 5 x 5 pixel region from the top-left corner of the original image.

```python
img_array = np.array(img)
print("Original Image (5x5 top-left region):")
print(img_array[0:5, 0:5])
```

c) Apply image negative transformation using the formula: s = 255 - r

```python
negative_array = 255 - img_array
```

d) Also print the corresponding  5 x 5 region from the negative image.

```python
print("\nNegative Image (5x5 top-left region):")
print(negative_array[0:5, 0:5])
```

e) Display both images

```python
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(img_array, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(negative_array, cmap='gray')
plt.title("Negative Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
